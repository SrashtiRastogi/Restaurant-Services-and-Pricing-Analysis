```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset  (1).csv')
df
```

## Analyze The Distribution of Aggregate Ratings and Determine The Most Common Rating Range

```python
#Find agregrate ratings

aggregate_ratings = df['Aggregate rating']
rating_counts = aggregate_ratings.value_counts().sort_index()
print(rating_counts)
```

```python
#Create Histogram

plt.hist(aggregate_ratings, bins = 20, edgecolor = 'black', color= 'lightpink')
plt.title("The Distribution of Aggregate Ratings")
plt.xlabel("Aggregate rating")
plt.ylabel("Frequency")
plt.show()
```

```python
#Find the most common rating range

most_common_rating_range = rating_counts.idxmax()
print("The most common rating range is:",most_common_rating_range)
```

## The Average Number Of Votes Received by Restaurants

```python
#Find the average number of votes received by restaurants

average_votes = df['Votes'].mean()
print("The average number of votes received by Restaurants is:",average_votes)
```
