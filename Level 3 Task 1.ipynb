```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset  (1).csv')
df
```

```python
# Display basic information about the dataset

df.info(), df.head()
```

## Analyze the Text Reviews to Identify the Most Common Positive and Negative Keywords

```python
# Count occurrences of rating text categories

rating_counts = df["Rating text"].value_counts()
print("\nMost Common Positive/Negative Keywords (Rating Text):")
print(rating_counts)
```

```python
# Count number of character

df['Review Length'] = df['Rating text'].astype(str).apply(len)

avg_len = df['Review Length'].mean()
print(f"\nAverage Length of Rating Text: {avg_len:.2f} characters")
```

```python
# Visualize the distribution of rating text categories

plt.figure(figsize=(10, 6))
sns.countplot(x='Rating text', data=df, order=['Not rated', 'Poor', 'Average', 'Good', 'Very Good', 'Excellent'], hue='Rating text', palette='viridis', legend=False)
plt.title('Frequency of Rating Texts (Positive vs Negative)')
plt.xlabel('Rating Text (Keywords)')
plt.ylabel('Count')
plt.show()
```

## Calculate the Average Length of Reviews and Explore if There is a Relationship Between Review Length and Rating.

```python
# Calculate the length of rating text

df["Rating text length"] = df["Rating text"].apply(lambda x: len(x.split()))
```

```python
# Compute the average length of rating text

average_length = df["Rating text length"].mean()
print(f"\nAverage Length of Rating Text: {average_length:.2f} characters")
```

```python
# visualize the relationship between rating text length and aggregate rating

plt.figure(figsize=(9, 6))
sns.boxplot(
    x='Rating text', 
    y='Review Length', 
    data=df, 
    order=['Not rated', 'Poor', 'Average', 'Good', 'Very Good', 'Excellent'], 
    hue='Rating text', 
    palette='pastel',
    legend=False
)
plt.title('Review Length vs Rating Text')
plt.xlabel('Rating Text')
plt.ylabel('Length of Text')
plt.show()
```
