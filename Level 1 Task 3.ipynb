```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset  (1).csv')
df
```

 ## The Distribution of Price Ranges Among The Restaurants

```python
#Price range counts

price_ranges_per_counts = df['Price range'].value_counts().sort_index()
print(price_ranges_per_counts)
```

```python
#Create bar chart


plt.figure(figsize=(8, 5))
colors = sns.color_palette('viridis', len(price_ranges_per_counts))
plt.bar(price_ranges_per_counts.index, price_ranges_per_counts.values, color=colors)
plt.title("The Distribution of Price Ranges Among The Restaurants")
plt.xlabel("Price Ranges")
plt.ylabel("Count")
plt.show()
```

```python
#Calculate total number of resturants

total_resturants = len(df['Price range'])
print("\nTotal Resturants :",total_resturants)
```

```python
#Calculate the percentage of resturants in each price range category

price_percentage = (price_ranges_per_counts / total_resturants) * 100

print("The Percentage of Resturants In Each price range Category:")
print(price_percentage)
```
