```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset  (1).csv')
df
```

## Identify the Restaurant with the Highest and Lowest Number of Votes

```python
# Identify the restaurant with the highet and lowest and number of votes

highest_votes = df.loc[df["Votes"].idxmax()]
lowest_votes = df.loc[df["Votes"].idxmin()]
```

```python
print(f"\nRestaurant with HIGHEST Votes: {highest_votes['Restaurant Name']}")
print(f"Votes Count: {highest_votes['Votes']}")
print(f"Rating: {highest_votes['Aggregate rating']}")

print(f"\nRestaurant with LOWEST Votes: {lowest_votes['Restaurant Name']}")
print(f"Votes Count: {lowest_votes['Votes']}")
print(f"Rating: {lowest_votes['Aggregate rating']}")
```

## Analyze if There is a Correlation Between the Number of Votes and the Rating of a Restaurant

```python
# Check correlation between votes and Aggregate rating

correlation = df["Votes"].corr(df["Aggregate rating"])
print(f"\nCorrelation between Votes and Rating: {correlation}")
```

```python
# Visualize the relationship using a scatter plot

plt.figure(figsize=(10, 5))
sns.scatterplot(x='Votes', y='Aggregate rating', data=df, alpha=0.5, color='purple')

plt.title('Relationship between Votes and Rating')
plt.xlabel('Number of Votes (Log Scale)')
plt.ylabel('Aggregate Rating')
plt.xscale('log') # Log scale makes the graph easier to read
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
