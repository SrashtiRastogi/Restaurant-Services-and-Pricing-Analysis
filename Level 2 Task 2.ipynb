```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset  (1).csv')
df
```

```python
# Drop missing values in 'Cuisines' column

df = df.dropna(subset=['Cuisines'])
```

## Identify The Most Common Combinations of Cuisines in The Dataset

```python
#Identify most common cuisine combinations

common_cuisine_combinations = df['Cuisines'].value_counts().head(10)
```

```python
#Display the top 10 most common cuisine combinations

print("Top 10 Most Common Cuisine Combinations")
print(common_cuisine_combinations)
```

```python
# Visualization top 10 Most Common Cuisines


plt.figure(figsize=(12, 6))
sns.barplot(x=common_cuisine_combinations.values, y=common_cuisine_combinations.index, hue=common_cuisine_combinations.index, legend=False, palette='Set2')
plt.title('Top 10 Most Common Cuisine Combinations')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine Combination')
plt.show()
```

## Determine if Certain Cuisine Combinations tend to Have Higher Ratings

```python
# Analyze cuisine combinations and their average ratings

cuisine_ratings = df.groupby("Cuisines")["Aggregate rating"].mean().nlargest(10)
```

```python
# Display the top 10 highest-rated cuisine combinations

print("\nTop 10 Highest-Rated Cuisine Combinations is:")
print(cuisine_ratings)
```

```python
# Visualization top 10 Highest Rated Cuisines


plt.figure(figsize=(11, 6))
sns.barplot(x=cuisine_ratings.values, y=cuisine_ratings.index, hue=cuisine_ratings.index, legend=False, palette='viridis')
plt.title('Top 10 Highest-Rated Cuisine Combinations')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine Combination')
plt.show()
```
