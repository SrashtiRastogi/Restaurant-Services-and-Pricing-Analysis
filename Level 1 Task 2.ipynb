```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dataset  (1).csv')
df
```

## Identify The City With The Highest Number Of Restaurants

```python
#City counts 

city_counts = df['City'].value_counts()
city_with_the_most_restaurants = city_counts.index[0]
print(f"The city with the highest number of restaurants is: {city_with_the_most_restaurants}")
```

## Calculate The Average Rating For Restaurants in Each City

```python
#Find aggegate ratings

average_rating_of_restaurants_in_each_city = df.groupby('City')['Aggregate rating'].mean()
print("The average rating for restaurants in each City")
print(average_rating_of_restaurants_in_each_city)
```

## City With The Highest Average Rating

```python
#Find average ratings

city_with_highest_average_rating = average_rating_of_restaurants_in_each_city.idxmax()
print(f"The city with the highest average rating is: {city_with_highest_average_rating}")
```

```python
plt.figure(figsize=(12, 8))
city_with_the_most_restaurants = city_counts.head(10)

sns.barplot(x=city_with_the_most_restaurants.index, y=city_with_the_most_restaurants.values, palette='viridis', hue=city_with_the_most_restaurants.index, legend=False)

plt.title('Top 10 Cities with Most Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45) 
plt.show()
```
